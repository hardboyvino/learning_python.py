```python
# import libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
```

```python
# read csv and drop unnecessary columns
df = pd.read_csv("../Boston.csv").drop("Unnamed: 0", axis=1)
df.head(3)
```

```python
# create new column for interactive item between lstat and age
df["lstat:age"] = df["lstat"] * df["age"]
df.head(3)
```

```python
# assign variables for X and y
X = df[["lstat", "age", "lstat:age"]]
y = df["medv"]
```

```python
# add the constant term
X = sm.add_constant(X)
```

```python
# regress and fit model
model = sm.OLS(y, X).fit()
```

```python
# print out summary
model.summary()
```
